```python
import urllib.request
html = urllib.request.urlopen(" http://python-data.dr-chuck.net/comments_282980.html")
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, "lxml")
print(soup.prettify())
```

```python
import urllib.request
html = urllib.request.urlopen(" http://python-data.dr-chuck.net/comments_282980.html")
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, "lxml")
tags = soup('span')
s = 0
for tag in tags:
    s += int(tag.contents[0])
print("Sum: ", s)
```
